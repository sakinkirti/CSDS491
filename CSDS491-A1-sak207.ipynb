1.1. Prove (5 pts)
$$ p(x,y|z) = p(x|z)p(y|x,z) $$
We can produce the product rule from $p(x,y|z)$ by first expanding it:
$$ p(x,y|z) = \frac{p(x, y, z)}{p(x)} $$
Then, we multiply the top and bottom by $p(x, z)$ which is equal to multiplying by 1:
$$ = \frac{p(x, y, z)}{p(x)} * \frac{p(x, z)}{p(x, z)} $$
Simplify:
$$ = \frac{p(x, z)}{p(z)} * \frac{p(x, y, z)}{p(x, z)} $$
Finally, we can apply the rule of conditional probability in reverse:
$$ = p(x|z)p(y|x, z) $$

1.2. Prove (5 pts)
$$ p(x|y,z) = \frac{p(y|x,z)p(x|z)}{p(y|z)} $$
We can prove this by first expanding it:
$$ p(x|y,z) = \frac{p(x, y, z)}{p(y,z)} $$
We can then normalize by dividing by $p(z)$
$$ = \frac{\frac{p(x,y,z)}{p(z)}}{\frac{p(y,z)}{p(z)}} $$
Simplify
$$ = \frac{\frac{p(x,y,z)}{x,z} * \frac{p(x,z)}{p(z)}}{p(y|z)} $$
Simplify
$$ = \frac{p(y|x, z)p(x|z)}{p(y|z)} $$

4.3 Perform Bernoulli trials where one of these views is correct.  Show how the posterior distribution of $\theta$ changes for each view for $n$=0, 1, 2, 5, 10, and 100.  Each view should have its own plot, but with the curves of the posterior after different numbers of trials overlayed. (5 pts)

```python
# my implementation of the bernoulli trials
import numpy as np
from scipy import stats

# helper function to calculate posterior
def calc_posterior(prior: float, y:int, n:int, a:float, b:float):
    """
    function to calculate posterior
    
    params:
    prior: float - the prior belief of theta
    y: int - the number of true outcomes
    n: int - the total number of trials
    a: float - alpha parameter for beta distribution
    b: float - beta parameter for beta distirbution
    """

    return stats.beta.pdf(prior, a+y, b+n-y)

def plot_posterior(a:float, b:float, theta:float, num_trails:list):
    """
    function to plot the posterior distribution
    
    params:
    a: float - the alpha param for beta distribution
    b: float - the beta param for beta distribution
    theta: float 0 - prior belief of theta
    num_trials: list - list of number of trials
    """

    prior = stats.bernoulli(theta)
    B = prior.rvs(max(num_trails))
    X = np.linspace(0, 1, 1000)

    for i in range(len(num_trails)):
        n = num_trails[i]
        y = sum(B[:n])
        posterior = calc_posterior(X, y, n, a, b)
        plt.plot(X, posterior, label = f'n = {n}')
    plt.legend()
    plt.xlabel("p(theta | y, n)")
    plt.ylabel("theta")
    plt.title(f"alpha = {a}, beta = {b}")

    plt.show()
```

```python
# test by creating several plots showing change of beliefs
trial_list = [0, 1, 2, 5, 10, 100]

plot_posterior(5, 1, 0.5, trial_list)
plot_posterior(1, 5, 0.5, trial_list)
plot_posterior(0.5, 0.5, 0.5, trial_list)
plot_posterior(5, 5, 0.5, trial_list)
plot_posterior(1, 1, 0.5, trial_list)
```

5.2 Derive the posterior distribution of $\lambda$ assuming a Gamma prior (usually defined with parameters $\alpha$ and $\beta$).  The posterior should have the form $p(\lambda | n, T, \alpha, \beta)$ where $T$ is the total duration of the observation period and $n$ is the number of events observed within that period. (5 pts)
<br>
$$
p(\lambda|n,T,\alpha,\beta) = \frac{p(n,T|\lambda,\alpha,\beta)p(\lambda|\alpha,\beta)p(\alpha,\beta)}{p(n,T|\alpha,\beta)p(\alpha,\beta)}
= \frac{p(n,T|\lambda,\alpha,\beta)\Gamma(\lambda;\alpha,\beta)}{p(n,T|\alpha,\beta)}
= \frac{p(n,T|\lambda,\alpha,\beta)\Gamma(\lambda;\alpha,\beta)}{\int_{0}^{\infty} p(n,T|\lambda,\alpha,\beta)p(\lambda|\alpha,\beta)d\lambda}
= \frac{ \frac{e^{-\lambda T}(\lambda{T})^n }{n!} \cdot \Gamma(\lambda; \alpha,\beta)}{\int_0^{\infty}(\frac{e^{-\lambda{T}}(\lambda{T})^n}{n!}  \cdot \Gamma (\lambda ; \alpha,\beta))d\lambda}
= \frac{ e^{-\lambda{T}}(\lambda T )^n \cdot \lambda^{\alpha-1} e^{-\beta \lambda}}{\int_0^{\infty} e^{-\lambda (\beta + T)}(T \lambda)^n  \lambda^{\alpha-1} \cdot d\lambda}
$$

5.3 Show that the Gamma distribution is a *conjugate prior* for the Poisson distribution, i.e. it is also a Gamma distribution, but defined by parameters $\alpha'$ and $\beta'$ that are functions of the prior and likelihood parameters. (5 pts)
<br>
$$
p(\lambda \mid n, T, \alpha, \beta) = \frac{ e^{-\lambda{T}}(\lambda T )^n \cdot \lambda^{\alpha-1} e^{-\beta \lambda}}{\int_0^{\infty} e^{-\lambda (\beta + T)}(T \lambda)^n  \lambda^{\alpha-1} \cdot d\lambda}
= \frac{\frac{(\beta + T)^{\alpha+n}T^n \lambda^{n}e^{-\lambda(\beta+T)}\lambda^{\alpha-1}}{\Gamma(a+n)}}{T^n \cdot \frac{\Gamma(a + n)}{\Gamma(a)}\frac{\beta^{\alpha}}{(T + \beta)^{a+n}}}{T^n \int_0^{\infty} \frac{(\beta+T)^{\alpha+n}\lambda^n e^{-\lambda(\beta+T)}}{\Gamma(a+n)}d\lambda}
= \frac{(\beta+T)^{\alpha+n}\lambda^{\alpha+n-1}e^{-\lambda(\beta+T)}}{\Gamma(a+n)}
= \Gamma(\lambda; \alpha+n, \beta+T)
$$
Therefore, the gamma distribution is the conjugate prior of the poisson distribution with $\alpha' = \alpha+n$ and $\beta' = \beta+n$
